```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# LAMMPS

### SNAP

```python
# X axis
num_nodes = [16, 32, 64, 128]

# Y axis
cray_mpich = [12.960, 25.168, 39.758, 75.861]
argonne_mpich = [12.913, 25.360, 39.604, 75.455]
openmpi = [12.861, 25.250, 39.552, 75.693]

# Set plot style and create a color palette
sns.set(style="whitegrid")
color_palette = sns.color_palette("colorblind", 3)

# Set the plot size
plt.figure(dpi=300)
fig, ax = plt.subplots(figsize=(16, 12))

# Increase font sizes
plt.rc('font', size=300)  # Adjusting the size for better visibility on a bar chart

# Set up the bar width and the positions for bars
bar_width = 0.25
r1 = range(len(num_nodes))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create bars
ax.bar(r1, cray_mpich, width=bar_width, color=color_palette[2], label='Cray MPICH')
ax.bar(r2, openmpi, width=bar_width, color=color_palette[0], label='Open MPI')
ax.bar(r3, argonne_mpich, width=bar_width, color=color_palette[1], label='Argonne MPICH')


# Set labels and title
ax.set_xlabel('Nodes')
ax.set_ylabel('Matom-GPU/s')
# ax.set_title('LAMMPS SNAP Potential ')

# Adjust the x-ticks to be centered among the group of bars for each node value
ax.set_xticks([r + bar_width for r in range(len(num_nodes))])
ax.set_xticklabels(num_nodes)  # Rotate x-axis labels for better visibility

# Set y ticks
ax.set_yticks(range(0, 81, 10))

# Add a legend and place it outside the plot area
legend = ax.legend(loc='upper left')

# Add grid for easier value estimation
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adjust the layout to prevent labels from being cut off
fig.tight_layout()

# Save plot
plt.savefig("figures/apps/lammps_snap.pdf", format="pdf", dpi=300, bbox_inches='tight', pad_inches=0.1)

# Display the plot
plt.show()
```
